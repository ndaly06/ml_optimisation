```python
# imports the required packages
import matplotlib as plt
import matplotlib.pyplot as plt
# inline plotting
%matplotlib inline
import missingno as msno
import numpy as np
import pandas as pd
import seaborn as sns
from seaborn import countplot
from matplotlib.pyplot import figure, show
```

```python
# reads in the model data to a pandas dataframe
model_data = pd.read_csv('/Users/nialdaly/Documents/ml_optimisation/data/pw_code_model_data.csv', index_col=False)

model_data.head()
```

## Assessing data quality

```python
# white fields indictate NA's
msno.matrix(model_data)
```

## Dataset analysis

```python
# counts the occurrence of datasets
dataset_count = model_data['Dataset'].value_counts()

# displays the top 10 datasets
dataset_count = dataset_count[:10]

dataset_count
```

```python
# plot showing the most frequent datasets
plt.figure(figsize=(14, 8))

sns.barplot(dataset_count.values, dataset_count.index, alpha=0.8)

plt.title('Dataset count (Top 10)')
plt.ylabel('Dataset', fontsize=12)
plt.xlabel('Count', fontsize=12)
```

## Model analysis

```python
# counts the occurrence of models
model_count = model_data['Model'].value_counts()

# displays the top 10 models
model_count = model_count[:10]

model_count
```

```python
# plot showing the most frequent datasets
plt.figure(figsize=(14, 8))

sns.barplot(model_count.values, model_count.index, alpha=0.8)

plt.title('Model count (Top 10)')
plt.ylabel('Model', fontsize=12)
plt.xlabel('Count', fontsize=12)
```
