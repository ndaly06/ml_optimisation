```python
# initialises an empty model data dataframe
model_data = pd.DataFrame(columns=['Task','Dataset','Model',
                          'Metric name','Metric value',
                          'Global rank', 'remove', 'paper_title', 'paper_path']
                )

#
for i in range(len(paper_titles)):
    
    #
    paper_models = tables_data[i]
    
    #
    paper_models['paper'] = paper_titles[i]
    paper_models['paper_path'] = paper_paths[i]
#     paper_models['paper_url'] = paper_urls[i]
    
    #
    model_data = pd.concat([model_data, paper_models], sort=True)

#     print(i)
```

```python
# resets the index on the model_data dataframe 
model_data = model_data.reset_index(drop = True)

# drops the remove column on the model_data dataframe
model_data = model_data.drop('remove', 1)
model_data.head()
```

```python
# reads in the model data to a pandas dataframe
model_data = pd.read_csv('/Users/nialdaly/Documents/ml_optimisation/data/pw_code_model_data.csv', index_col=False)

model_data.head()
```

```python
# #
# model_data.rename(columns={'paper':'paper_title'}, inplace=True)

# # drops the remove column on the model_data dataframe
# model_data = model_data.drop('paper_url', 1)
```

```python
# lowercases the model_data column names
model_data.columns = map(str.lower, model_data.columns)

# replaces whitespace in the model data column names with underscores
model_data.columns = model_data.columns.str.replace(' ', '_')

model_data.head()
```

```python
# reads in the paper_urls data to a pandas dataframe
paper_urls = pd.read_csv('/Users/nialdaly/Documents/ml_optimisation/data/pw_code_paper_urls.csv', index_col=False)

paper_urls.head()
```

```python
# merges the model_data and paper_urls dataframes based on the paper_title column pandas
model_data = model_data.merge(paper_urls, on="paper_title", how = 'inner')

model_data.head()
```

```python
# exports the model_data dataframe to csv
model_data.to_csv('/Users/nialdaly/Documents/ml_optimisation/data/pw_code_model_data.csv', index = False)
```
